## 4. Visualize Training Results

After training is complete, the best model weights (`best.pt`) are saved. We can now load this fine-tuned model and use it to make predictions on new images. Let's run it on an image from the validation set to see how it performs.

```python
import glob
import os
import cv2
import matplotlib.pyplot as plt

try:
    latest_run_dir = max(glob.glob(os.path.join('runs', 'detect', 'train*')), key=os.path.getctime)
    model_path = os.path.join(latest_run_dir, 'weights', 'best.pt')
    print(f"Found latest model at: {model_path}")
except ValueError:
    print("Error: No training directories found. Please run the training cell first.")
    model_path = None

if model_path and os.path.exists(model_path):
    best_model = YOLO(model_path)
    test_image_paths = glob.glob('datasets/coco128/images/train2017/*.jpg') 

    if not test_image_paths:
        print("No test images found in 'datasets/coco128/images/test/'. Please verify the path.")
    else:
        print(f"Found {len(test_image_paths)} test images.")

        output_dir = 'runs/detect/test_inferences'
        os.makedirs(output_dir, exist_ok=True)
        print(f"Saving inference results to: {output_dir}")

        for i, image_path in enumerate(test_image_paths[:1]):
            print(f"\n--- Processing: {os.path.basename(image_path)} ---")

            # Perform prediction
            results = best_model.predict(source=image_path)
            
            # The .plot() method returns a BGR numpy array with detections (perfect for OpenCV)
            result_plotted_bgr = results[0].plot()
            
            result_rgb = cv2.cvtColor(result_plotted_bgr, cv2.COLOR_BGR2RGB)

            plt.figure(figsize=(10, 8))
            plt.imshow(result_rgb)
            plt.title(f"Detection Result - {os.path.basename(image_path)}")
            plt.axis('off')
            plt.show()

            save_image_path = os.path.join(output_dir, f"detected_{os.path.basename(image_path)}")
            cv2.imwrite(save_image_path, result_plotted_bgr)
            print(f"Saved detected image to: {save_image_path}")

        cv2.destroyAllWindows() 
        print("\nAll processing complete.")

else:
    print("Model not found or path is incorrect.")
```
